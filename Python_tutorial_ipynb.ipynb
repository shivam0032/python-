**Python DATA TYPE**

Python data types are classifications that tell us what kind of values a variable can hold. They determine the operations that can be performed on the data and the way it's stored in memory. Here are some of the most common built-in Python data types
Number (Integer): Whole numbers without a fractional part.

**String:** A sequence of characters, used for text.   
**Float:** Numbers with a decimal point.
List: An ordered, changeable collection of items.      
**Tuple:** An ordered, unchangeable collection of items.     
**Range:** An immutable sequence of numbers, often used for looping.    
**Dictionary:** An unordered, changeable collection of key-value pairs.   
**Boolean**: Represents truth values, either True or False.  
**Set:** An unordered, changeable collection of unique items.  
**None:** A special data type that represents the absence of a value.  
**Bytes:** A sequence of bytes, often used for binary data.   
**Complex:** Numbers with a real and an imaginary part.

**FUNCTION**

A function in Python is like a mini-program that performs a specific task. You give it a name, write some code inside it, and then you can call that name whenever you want to run that specific code. This helps you organize your code and reuse it easily.  
**Built-in Functions:** These are functions that are pre-defined in Python and are always available for use. You don't need to import anything to use them. Examples include print(), len(), type(), input(), range(), sum(), etc.

**User-defined Functions:** These are functions that you, the programmer, create to perform specific tasks. They allow you to organize your code, make it reusable, and improve readability. You define them using the def keyword.

**Lambda Functions **(Anonymous Functions)
These are small, single-expression functions that don't have a formal name. They are defined using the lambda keyword and are typically used for short, one-time operations. They are often used as arguments to higher-order functions like map(), filter(), and sorted().
