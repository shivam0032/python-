```python
import pandas as pd
import numpy as np

data={
    'user_id':[101,102,103,104,105],
    'country':['India','Usa','Uk','Canada','Australia'],
    'feedback_rating':[5,4,3,2,1],
    'usage_data':['2025-07-16','2025-07-15','2025-07-14','2025-07-13','2025-07-12'],
    'queries_handled':[10,11,12,13,14]
    }

df = pd.DataFrame(data)
np.random.seed(42) # for reproducibility
df['Session Length'] = np.random.randint(5, 60, size=len(df))

if 'Session Length' in df.columns and 'country' in df.columns:
    country_session_length = df.groupby('country')['Session Length'].mean().reset_index()
    print("\nCountry-Wise Average Session Length:")
    print(country_session_length)
else:
    print("'Session Length' or 'country' column not found.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

if 'feedback_rating' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(x='feedback_rating', data=df, palette='viridis')
    plt.title('Distribution of Feedback Ratings')
    plt.xlabel('Feedback Rating')
    plt.ylabel('Count')
    plt.show()
else:
    print("'feedback rating")
```

```python
if 'usage_data' in df.columns:
    df['usage_data'] = pd.to_datetime(df['usage_data'])
    if 'Session Length' not in df.columns:

        np.random.seed(42)
        df['Session Length'] = np.random.randint(5, 60, size=len(df))
    if 'Session Length' in df.columns:
        daily_avg_session = df.groupby(df['usage_data'].dt.date)['Session Length'].mean().reset_index()

        # Plot the time series
        plt.figure(figsize=(6, 6))
        sns.lineplot(x='usage_data', y='Session Length', data=daily_avg_session, marker='o')
        plt.title('Time Series of Average Session Length')
        plt.xlabel('Date')
        plt.ylabel('Average Session Length')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    else:
        print("'Session Length' column not found after creating dummy data.")
else:
    print("'usage_data' column not found.")
```

```python

numerical_cols = ['feedback_rating', 'queries_handled']
correlation_matrix = df[numerical_cols].corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
